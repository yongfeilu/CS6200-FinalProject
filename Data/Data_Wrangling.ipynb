```python
from elasticsearch import Elasticsearch
import os
import re
```

```python
# create index
# indexer source code
es = Elasticsearch("http://localhost:9200", request_timeout = 5000)
if es.ping():
    # Delete index /ap_dataset if it already exists
    es.indices.delete(index="listing_dataset", ignore=[400, 404])
    print("Deleted index listing_dataset if it already existed.")

    # Retrieve the names of all files to be indexed in folder ./AP_DATA/ap89_collection of the cwd.
#     for dir_path, dir_names, file_names in os.walk("./AP_DATA/ap89_collection"):
#         allfiles = [os.path.join(dir_path, filename).replace("\\", "/") for filename in file_names if (filename != "readme")]

    # Regular expressions to extract data from the corpus
    doc_regex = re.compile("<DOC>.*?</DOC>", re.DOTALL)
    docno_regex = re.compile("<DOCID>.*?</DOCID>")
    text_regex = re.compile("<DESC>.*?</DESC>", re.DOTALL)
    comm_regex = re.compile("<NEIG>.*?</NEIG>", re.DOTALL)
    lat_regex = re.compile("<LAT>.*?</LAT>", re.DOTALL)
    long_regex = re.compile("<LONGT>.*?</LONGT>", re.DOTALL)
    pic_regex = re.compile("<PIC>.*?</PIC>", re.DOTALL)
    name_regex = re.compile("<Name>.*?</Name>", re.DOTALL)
    # For all the files, read each file and fetch the documents. From the documents fetch the doc number and text
#     for file in allfiles:
    with open("./listings_doc.txt", 'r', encoding='ISO-8859-1') as f:
        filedata = f.read()
        result = re.findall(doc_regex, filedata)  # Match the <DOC> tags and fetch documents

        for document in result:
            # Retrieve contents of DOCNO tag
            docno = re.findall(docno_regex, document)[0].replace("<DOCID>", "").replace("</DOCID>", "").strip()
            # Retrieve contents of TEXT tag
            texts = "".join(re.findall(text_regex, document)).replace("<DESC>", "").replace("</DESC>", "")\
                    .replace("\n", " ")
            comm = re.findall(comm_regex, document)[0].replace("<NEIG>", "").replace("</NEIG>", "").strip()
            lat = re.findall(lat_regex, document)[0].replace("<LAT>", "").replace("</LAT>", "").strip()
            long = re.findall(long_regex, document)[0].replace("<LONGT>", "").replace("</LONGT>", "").strip()
            pic = re.findall(pic_regex, document)[0].replace("<PIC>", "").replace("</PIC>", "").strip()
            name = re.findall(name_regex, document)[0].replace("<Name>", "").replace("</Name>", "").strip()
            es.index(index="listing_dataset",
                     doc_type="document",
                     id= docno,
                     body={
                         "docno": docno,
                         "name":name,
                         "neighborhood" : comm,
                         "latitude" : lat,
                         "longitude" : long,
                         "pic_url": pic,
                         "text_content": texts
                     })
            print("Completed:{0}".format(docno))

    print("Indexing completed")

    es.indices.refresh(index="listing_dataset")
    print("Index refreshed.")

else:
    print("Elastic search is not running")
```

```python
def search(query):
    es = Elasticsearch(["http://localhost:9200/"], timeout=5000)
    result = es.search(index="listing_dataset",
                       doc_type="document",
                       body={
                           "query": {
                                "match": {
                                    "text_content": query
                                }
                           },
                           
                       })
  
    
    result_DF = result['hits']['total']
    result_hits =  result['hits']['hits']
    print(result_DF)
    print(result_hits)
    return result_DF, result_hits
```

## Wrangle Review data

```python
reviews.isnull().sum()
```

```python
# drop rows with null comments
# some reviews will also have the prototypical 'Host canceled this booking x days before', drop rows
# misc preprocessing: remove numeric, -


reviews.dropna(subset =['comments'], how='any', axis=0, inplace=True)

index_canceled = reviews[ reviews['comments'].str.match('The host canceled this reservation')].index
reviews.drop(index_canceled, inplace=True)

index_dash = reviews[ reviews['comments'].str.match('-')].index
reviews.drop(index_dash, inplace=True)

alphanumeric = lambda x: re.sub('\w*\d\w*', ' ', x)
reviews['comments'] = reviews['comments'].map(alphanumeric)
```

```python
reviews.reset_index(inplace=True, drop=True)
```

```python
# some reviews seem to be in other languages, have to detect the language and remove all non-English comments
# langdetect module can do that for us

num_errors = 0
err = []

for index, row in reviews.iterrows():
    try:
        language = detect(row['comments']) #(" ".join(row.split())) # row.strip())
    except:
        num_errors += 1
        err.append(index)
        print("This row throws an error:", row['comments'])
```

```python
print('{:.2f}% of entries for language identification throw errors'.format(num_errors/reviews['comments'].shape[0] * 100))
```

```python
reviews.drop(err, axis=0, inplace=True)
reviews.reset_index(inplace=True, drop=True)
```

```python
# and create new col 'language' with detected language
reviews['language'] = reviews['comments'].apply(detect)
reviews.head()
```

```python
reviews['language'].value_counts().plot.bar()
```

```python
# create a copy in case we eventually want to try translating the non-English rows instead of dropping them
reviews_en = reviews.copy(deep=True)
```

```python


# isolate all non-en entries
index_nonen = reviews_en[~reviews_en['language'].str.match('en')].index

print('{:.2f}% of all entries are not in English'.format(len(index_nonen)/reviews_en.shape[0]))
```

```python
reviews_en.drop(index_nonen, inplace=True)
```

```python
analyzer = SentimentIntensityAnalyzer()

# test it out on first comment

def sentiment_analyzer_scores(comment):
    score = analyzer.polarity_scores(comment)
    return(pd.DataFrame.from_dict(score, orient='index'))

sentiment_analyzer_scores(reviews_en['comments'][0])
```
